### Scraping Daily Stock Data from Alpha Vantage

```python
import requests
import pandas as pd
import time

# Alpha Vantage API key
API_KEY = '4X5GDXTRCSG5GF63'

# List of stock symbols to scrape data from
stocks = ['AAPL', 'GOOGL', 'AMZN', 'MSFT', 'TSLA']

url = "https://www.alphavantage.co/query?function=TIME_SERIES_DAILY&symbol={}&apikey={}"

# Function to get stock data
def get_stock_data(symbol, api_key):
    try:
        response = requests.get(url.format(symbol, api_key))
        response.raise_for_status()  # Raise an exception for HTTP errors
        data = response.json()
        
        print(f"Response for {symbol}: {data}")
        
        if 'Error Message' in data:
            print(f"Error for {symbol}: {data['Error Message']}")
            return []
        
        # Extract only the time series data
        time_series = data.get('Time Series (Daily)', {})
        print(f"Time Series Data for {symbol}: {time_series}")
        
        stock_info = []
        for date, values in time_series.items():
            stock_info.append({
                'Date': date,
                'Stock': symbol,
                'Open': values['1. open'],
                'High': values['2. high'],
                'Low': values['3. low'],
                'Close': values['4. close'],
                'Volume': values['5. volume']
            })
        return stock_info
    except requests.RequestException as e:
        print(f"Request error for {symbol}: {e}")
        return []

# Scrape stock data for each company
all_stock_data = []
for stock in stocks:
    stock_data = get_stock_data(stock, API_KEY)
    all_stock_data.extend(stock_data)
    time.sleep(12)  # To avoid exceeding API rate limits
all_stock_data
```

# Price and Volume trend 

```python
import mplfinance as mpf
import matplotlib.pyplot as plt

def plot_stock(stock_symbol):
    # Filter the DataFrame for the selected stock
    stock_data = df[df['Stock'] == stock_symbol]
    
    if stock_data.empty:
        print(f"No data available for stock: {stock_symbol}")
        return
    
    # Create the candlestick chart
    mpf.plot(stock_data, type='candle', style='charles', 
             title=f"{stock_symbol} Candlestick Chart", 
             ylabel='Price (USD)', 
             volume=True, 
             mav=(3,6,9))  # Moving averages for visual guidance (optional)
    
plot_stock('AAPL')    # Stock: ['AAPL', 'AMZN', 'GOOGL', 'MSFT', 'TSLA']
```

```python
import mplfinance as mpf
import matplotlib.pyplot as plt
import seaborn as sns

def Price_Volatility(Stock_name):

    Stock_Data2 = df[df['Stock'] == Stock_name]

    if Stock_Data2.empty:
        print(f"No data available for stock: {Stock_name}")
        return
        
    Stock_Data2['Volatility'].plot(figsize=(8, 5), title="Stock Price Volatility",color = '#006400')
    plt.xlabel('Date')
    plt.ylabel('Volatility (15-day rolling std. deviation)')
    plt.show()

Price_Volatility('AAPL')    # Stock: ['AAPL', 'AMZN', 'GOOGL', 'MSFT', 'TSLA']
```

# Trading Volume Analysis
<!-- Visualize the daily trading volume to observe trends in trading activity. A significant increase in trading volume often accompanies price breakouts or shifts in trends. -->

```python
def Trading_Volume(stock_name):

    Stock_Data3 = df[df['Stock'] == stock_name]

    if Stock_Data3.empty:
        print(f"No data available for stock: {stock_name}")
        return
        
    fig, ax1 = plt.subplots(figsize=(8, 6))
    
    ax1.plot(Stock_Data3.index, Stock_Data3['Close'], color='b', label='Closing Price')
    ax1.set_xlabel('Date')
    ax1.set_ylabel('Closing Price (USD)', color='b')
    
    ax2 = ax1.twinx()
    ax2.bar(Stock_Data3.index, Stock_Data3['Volume'], alpha=0.4, color='orange', label='Volume')
    ax2.set_ylabel('Volume', color='#ffd700')
    
    plt.title("Closing Price vs. Volume")
    plt.grid(visible=None)
    plt.show()

Trading_Volume('AAPL')     # Stock: ['AAPL', 'AMZN', 'GOOGL', 'MSFT', 'TSLA']
```

# Relative Strength Index (RSI)
<!-- The RSI is a momentum indicator used to identify overbought or oversold conditions in a stock. It oscillates between 0 and 100, with values above 70 indicating overbought conditions and below 30 indicating oversold conditions. -->

```python
def calculate_rsi(data, window=14):
    
    delta = data['Close'].diff(1)
    gain = (delta.where(delta > 0, 0)).rolling(window=window).mean()
    loss = (-delta.where(delta < 0, 0)).rolling(window=window).mean()

    RS = gain / loss
    RSI = 100 - (100 / (1 + RS))
    return RSI

df['RSI'] = calculate_rsi(df)
```

```python
def RSI_vis(Stock_Name):

    Stock_Data4 = df[df['Stock'] == Stock_Name]

    if Stock_Data4.empty:
        print(f"No data available for stock: {Stock_Name}")
        return
    Stock_Data4['RSI'].plot(figsize=(10, 6), title="Relative Strength Index (RSI)")
    plt.axhline(70, linestyle='--', alpha=0.5, color='red')
    plt.axhline(30, linestyle='--', alpha=0.5, color='green')
    # plt.grid(visible=None)
    plt.show()

RSI_vis('AAPL')     # Stock: ['AAPL', 'AMZN', 'GOOGL', 'MSFT', 'TSLA']
```
